```python
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
# FFT
N = 1000
f = open("tdgl_result.dat", "r")
i = 0
xx = np.zeros(N)
yy = np.zeros(N)
for row in f:
  if i == 0:
    # Retrieve parameters
    N, tmax, dx, dt, seed, A, Thalf = row.split(" ")
    N = int(N)
    tmax = float(tmax)
    dx = float(dx)
    dt = float(dt)
    seed = int(seed)
  else:
    # Plot data
    xx[i-1], yy[i-1] = row.split(" ")  
  i = i + 1


plt.plot(xx,yy)
plt.ylim(-1,1)
plt.title("N = " + str(N) + " t = " + str(tmax) + "\n dx = " + str(dx) + " dt = " + str(dt))
if (os.path.isdir(str(seed))) == False:
  os.mkdir(str(seed))
plt.savefig(str(seed) + "//t " + str(tmax) + ".png")
```
